Write a function 'mysum()' to get the sum of the values in the input list xs.  Don't use Python function sum().

```python
def mysum(xs):
    """ Return the sum of the values in xs. """
    
    total = 0
    for x in xs:
        total += x
    return total
```

Write a function 'avg()' to get the average value of input list xs.  Raise a ValueError if list xs has length 0.

```python
def avg(xs):
    """ Return the average of the values in list xs. """
    
    if len(xs) == 0:
        raise ValueError('avg() argument is an empty list')
        
    return sum(xs)/len(xs)
```

Write a function 'sum_err()' to get the sum of the squared difference between each value in xs and the average value of xs.

For example, `sum_err([1,3,5])` should be 8, because `(1-3)**2`, `(3-3)**2`, and `(5-3)**2` is 8.


```python
def sum_err(xs):
    """ Return the sum of the squared differences between between
    the values in xs and the average value of xs.
    """
    
    esum = 0
    xs_avg = avg(xs)
    for x in xs:
        esum += (x - xs_avg)**2
        
    return esum
```

Test your function sum_error() on [1,3,5].

```python
sum_err([1,3,5])
```

Test your function on [0, 3, 8].

```python
sum_err([0, 3, 8])
```
